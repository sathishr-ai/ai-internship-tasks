```python
from text_generator import load_model_and_tokenizer, generate_text

# Load the model and tokenizer
print("Loading GPT-2...")
tokenizer, model = load_model_and_tokenizer()
print("Model loaded successfully!")
```

```python
# Example 2: Creative Writing Prompt
prompt2 = "In a world where magic is replaced by technology,"
generated2 = generate_text(prompt2, tokenizer, model, max_length=150)
print("Prompt:", prompt2)
print("Generated Text:\n", generated2)
```
